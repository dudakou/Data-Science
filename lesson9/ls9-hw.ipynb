## Корреляции

```python
correlation_matrix = df.corr().round(1)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

Оставить только положительные корреляции выше 0,6:

```python
correlation_matrix.where(correlation_matrix > 0.6)
```

## Корреляции со стоимостью жилья

```python
medv_correlations = correlation_matrix['medv'].sort_values(ascending=False)
print(medv_correlations)
```

## Как количество комнат влияет на стоимость жилья

```python
rm_sorted_data = df.sort_values('rm')

plt.figure(figsize=(10, 6))
plt.plot(rm_sorted_data['rm'], rm_sorted_data['medv'], marker='o', linestyle='-', color='b')
plt.title('Зависимость MEDV от RM')
plt.xlabel('Кол-во комнат в доме')
plt.ylabel('Медианная стоимость жилья')
plt.grid(True)
plt.show()
```

Стоимость жилья по категориям кол-ва комнат:

```python
df['rm_category'] = pd.cut(df['rm'], bins=[0, 5, 7, 9], 
                           labels=['Small (<5)', 'Medium (5-7)', 'Large (7+)'])
rm_to_medv = df.groupby('rm_category')['medv'].agg(['mean', 'count'])
print(rm_to_medv)
```

```python
category_counts = df['rm_category'].value_counts()
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Распределение домов по категориям количества комнат')
plt.xlabel('Категории количества комнат')
plt.ylabel('Количество домов')
plt.show()
```

## Как доля низкого соцстатуса населения влияет на стоимость жилья

```python
lstat_sorted_data = df.sort_values('lstat')

plt.figure(figsize=(10, 6))
plt.plot(lstat_sorted_data['lstat'], lstat_sorted_data['medv'], marker='o', linestyle='-', color='b')
plt.title('Зависимость MEDV от LSTAT')
plt.xlabel('% низкого социального статуса населения')
plt.ylabel('Медианная стоимость жилья')
plt.grid(True)
plt.show()
```

Стоимость жилья по категориям доли низкого соцстатуса населения:

```python
df['lstat_category'] = pd.cut(df['lstat'], bins=[0, 5, 10, 20, 50], 
                           labels=['0-5%', '5-10%', '10-20%', '20+%'])
lstat_to_medv = df.groupby('lstat_category', observed=False)['medv'].agg(['mean', 'count'])
print(lstat_to_medv)
```

## Как доля земель, занятых предприятиями, складами, офисами влияет стоимость жилья

```python
indus_sorted_data = df.sort_values('indus')

plt.figure(figsize=(10, 6))
plt.plot(indus_sorted_data['indus'], indus_sorted_data['medv'], marker='o', linestyle='-', color='b')
plt.title('Зависимость MEDV от INDUS')
plt.xlabel('% non-retail business')
plt.ylabel('Медианная стоимость жилья')
plt.grid(True)
plt.show()
```

Стоимость жилья по категориям:

```python
df['indus_category'] = pd.cut(df['indus'], bins=[0, 5, 15, 30], 
                           labels=['0-5%', '5-15%', '15%+'])
indus_to_medv = df.groupby('indus_category', observed=False)['medv'].agg(['mean', 'count'])
print(indus_to_medv)
```

## Как налоговая ставка влияет на стоимость жилья

```python
tax_sorted_data = df.sort_values('tax')

plt.figure(figsize=(10, 6))
plt.plot(tax_sorted_data['tax'], tax_sorted_data['medv'], marker='o', linestyle='-', color='b')
plt.title('Зависимость MEDV от TAX')
plt.xlabel('Ставка налога на недвижимость')
plt.ylabel('Медианная стоимость жилья')
plt.grid(True)
plt.show()
```

```python
df['tax_category'] = pd.cut(df['tax'], bins=[0, 300, 450, 800], 
                           labels=['<$300 per $10.000', '$300-450 per $10.000', '$450+ per $10.000'])
tax_to_medv = df.groupby('tax_category', observed=False)['medv'].agg(['mean', 'count'])
print(tax_to_medv)
```

## Как на стоимость жилья влияет ср.кол-во учеников, которые приходятся на 1 учителя

```python
ptratio_sorted_data = df.sort_values('ptratio')

plt.figure(figsize=(10, 6))
plt.plot(ptratio_sorted_data['ptratio'], ptratio_sorted_data['medv'], marker='o', linestyle='-', color='b')
plt.title('Зависимость MEDV от PTRATIO')
plt.xlabel('Сколько учеников приходится на 1 учителя')
plt.ylabel('Медианная стоимость жилья')
plt.grid(True)
plt.show()
```

```python
df['ptratio_category'] = pd.cut(df['ptratio'], bins=[0, 14, 18, 20, 30], 
                           labels=['до 14 учеников на 1 учителя', '14-18 учеников на 1 учителя',
                                   '18-20 учеников на 1 учителя', '20+ учеников на 1 учителя'])
ptratio_to_medv = df.groupby('ptratio_category', observed=False)['medv'].agg(['mean', 'count'])
print(ptratio_to_medv)
```
